3. Выведите колонки, у которых медиана равна минимальному значению (название колонки).
Напишите какой вывод можно сделать по данным в этих колонках.

###### Вывод:
Оба признака не сбалансированы (кол-во одного значения данного признака по всей выборке сильно приобладает над остальными (>50% от всех наблюдений)).
В то-же время признак CHAS - бинарный, в распределении бинарного признака - медиана всегда будет равна либо минимальному либо максимальному значению (за исключением случая когда кол-во наблюдений четное и кол-во разных значений (1 и 0) по выборке равно друг другу, тогда медиана равна их среднему (0.5)). Но в любом случае признак CHAS то-же не сбалансирован.

```python
pd.read_sql(
    """
    WITH 
        crim_cond AS (SELECT 'CRIM' as colname, avg(crim_d)=min(crim) AS cond FROM
            (SELECT crim AS crim_d FROM boston ORDER BY crim
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
  
        zn_cond AS (SELECT 'ZN' as colname, avg(zn_d)=min(zn) AS cond FROM
            (SELECT zn AS zn_d FROM boston ORDER BY zn
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
 
        indus_cond AS (SELECT 'INDUS' as colname, avg(indus_d)=min(indus) AS cond FROM
            (SELECT indus AS indus_d FROM boston ORDER BY indus
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
 
        chas_cond AS (SELECT 'CHAS' as colname, avg(chas_d)=min(chas) AS cond FROM
            (SELECT chas AS chas_d FROM boston ORDER BY chas
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
           
        nox_cond AS (SELECT 'NOX' as colname, avg(nox_d)=min(nox) AS cond FROM
            (SELECT nox AS nox_d FROM boston ORDER BY nox
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),

        rm_cond AS (SELECT 'RM' as colname, avg(rm_d)=min(rm) AS cond FROM
            (SELECT rm AS rm_d FROM boston ORDER BY rm
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
            
        age_cond AS (SELECT 'AGE' as colname, avg(age_d)=min(age) AS cond FROM
            (SELECT age AS age_d FROM boston ORDER BY age
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
        
        dis_cond AS (SELECT 'DIS' as colname, avg(dis_d)=min(dis) AS cond FROM
            (SELECT dis AS dis_d FROM boston ORDER BY dis
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
            
        rad_cond AS (SELECT 'RAD' as colname, avg(rad_d)=min(rad) AS cond FROM
            (SELECT rad AS rad_d FROM boston ORDER BY rad
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
            
        tax_cond AS (SELECT 'TAX' as colname, avg(tax_d)=min(tax) AS cond FROM
            (SELECT tax AS tax_d FROM boston ORDER BY tax
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),    
            
        ptratio_cond AS (SELECT 'PTRATIO' as colname, avg(ptratio_d)=min(ptratio) AS cond FROM
            (SELECT ptratio AS ptratio_d FROM boston ORDER BY ptratio
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),
            
        b_cond AS (SELECT 'B' as colname, avg(b_d)=min(b) AS cond FROM
            (SELECT b AS b_d FROM boston ORDER BY b
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),

        lstat_cond AS (SELECT 'LSTAT' as colname, avg(lstat_d)=min(lstat) AS cond FROM
            (SELECT lstat AS lstat_d FROM boston ORDER BY lstat
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston),

        medv_cond AS (SELECT 'MEDV' as colname, avg(medv_d)=min(medv) AS cond FROM
            (SELECT medv AS medv_d FROM boston ORDER BY medv
            LIMIT 2 OFFSET (SELECT (count() - 1) / 2 FROM boston)), boston)
            
            
            
    SELECT colname FROM (
        SELECT * FROM crim_cond
        UNION ALL
        SELECT * FROM zn_cond
        UNION ALL
        SELECT * FROM indus_cond
        UNION ALL
        SELECT * FROM chas_cond
        UNION ALL
        SELECT * FROM nox_cond
        UNION ALL
        SELECT * FROM rm_cond
        UNION ALL
        SELECT * FROM age_cond
        UNION ALL
        SELECT * FROM rad_cond
        UNION ALL
        SELECT * FROM tax_cond
        UNION ALL
        SELECT * FROM ptratio_cond
        UNION ALL
        SELECT * FROM b_cond
        UNION ALL
        SELECT * FROM lstat_cond
        UNION ALL
        SELECT * FROM medv_cond
    )
    WHERE cond=1
    
    """, con)
```

4. Выведите разницу между средним количеством комнат(RM) в домах с самой дорогой стоимостью(MEDV) и 25 самыми дешевыми домами. 
Аналогично по 50, 100, 200, 300 самыми дешевыми домами. (кол-во домов(25,50,100,200,300), среднее кол-во комнат в них, среднее кол-во комнат в самых дорогих, разница).
Напишите влияет ли кол-во комнат на стоимость и как сильно.

##### Вывод:

Влияет - видна положительная корреляция между данными признаками. 

```python
pd.read_sql(
    """
    WITH 
        h AS 
            (SELECT 
                avg(rm) AS avg_high
            FROM boston
            GROUP BY medv
            ORDER BY medv DESC
            LIMIT 1),
        min_25 AS 
            (SELECT 
                25 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS diff
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 25), h),
        min_50 AS 
            (SELECT 
                50 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS diff
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 50), h),
        min_100 AS 
            (SELECT 
                100 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS diff
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 100), h),
        min_200 AS 
            (SELECT 
                200 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS diff 
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 200), h),
        min_300 AS 
            (SELECT 
                300 AS n_houses,
                avg(rm) AS rm_avg,
                avg_high AS rm_avg_h,
                avg_high-avg(rm) AS diff
            FROM
                (SELECT rm, medv
                FROM boston
                ORDER BY medv ASC
                LIMIT 300), h)
    
    SELECT * FROM min_25
    UNION ALL
    SELECT * FROM min_50
    UNION ALL
    SELECT * FROM min_100
    UNION ALL
    SELECT * FROM min_200
    UNION ALL
    SELECT * FROM min_300
            
    """,
    con,
)
```

5. Выведите ранги значений колонки LSTAT(процент населения с более низким статусом) в домах с самой дорогой стоимостью (значение LSTAT, стоимость, ранг). Напишите какой вывод можно сделать по этим данным.

###### Вывод:
В домах с самой дорогой стоимостью (MEDV) разброс процента населения с более низким статусом (LSTAT) находится в первой половине всех значений LSTAT (т.е. с преимущественно низким значением процента населения с более низким статусом (LSTAT))

Итого - в самых дорогих домах процент населения с более низким статусом ожидаемо низок (распределение имеет меньший разброс значений и сдвинуто в сторону минимального значения относительно распределения всех значений LSTAT)

```python
pd.read_sql(
    """
    WITH ranked AS
        (SELECT 
            lstat,
            medv,
            RANK () OVER ( 
            ORDER BY lstat ASC
            ) lstat_rank
        FROM boston)
    SELECT 
        lstat,
        medv*1000 AS avg_price,
        lstat_rank
    FROM ranked
    WHERE medv = (SELECT max(medv) FROM boston)
    """,
    con,
)
```

6. Выведите среднюю стоимость домов граничащих с рекой(CHAS) и нет (граничит/не граничит, стоимость)

```python
pd.read_sql(
    """
    WITH 
        h_river AS 
            (SELECT 
                medv,
                chas
            FROM boston
            WHERE chas = 1),
        h_common AS 
            (SELECT 
                medv,
                chas 
            FROM boston
            WHERE chas = 0)
            
    SELECT 
        avg(medv)*1000 AS avg_price,
        chas
    FROM h_common
    UNION ALL
    SELECT 
        avg(medv)*1000 AS avg_price,
        chas
    FROM h_river
    """,
    con,
)
```
